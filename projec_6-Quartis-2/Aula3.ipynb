```python
import  pandas as pd
```

```python
df = pd.read_csv('dados_limpos.csv')
df
```

## Aplicar uma fórmula para calcular novas colunas

Isolar as colunas `city-mpg`	e `highway-mpg`

```python
df['city-kmL'] = df['city-mpg'] * 0.425144
df['highway-kmL'] = df['highway-mpg'] * 0.425144
df
```

Limpar as colunas antigas que não me interessam mais

```python
df.drop(['city-mpg', 'highway-mpg'], axis=1, inplace=True)
df
```
