```python
# Carregar o dataset com as colunas definidas
# Os arquivos .data sao um csv, os arquivos .name aom um .txt
df = pd.read_csv('/content/drive/MyDrive/Aula 2/automobile/imports-85.data', header=None)
df.columns = colunas
df
```

```python
'''No arquivo da base de dados Automobile, h√° uma descri√ß√£o detalhada das informa√ß√µes armazenadas. Neste arquivo, podemos identificar as colunas que possuem valores faltantes.
Com base nessas colunas, faremos a substitui√ß√£o dos valores ausentes,
representados por "?", por "None".'''
```

```python
# **1. Substituindo "?" por "None" nas colunas com valores faltantes**
# Identificando as colunas que possuem valores faltantes em uma lista

columns_with_missing_values = ['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price']
print(df)

# Substituindo "?" por "None" nas colunas identificadas
for column in columns_with_missing_values:
  df[column] = df[column].replace('?', None)

# Visualizando a mudan√ßa
print(df) #" Ver DataFrame ap√≥s a substitui√ß√£o de '?' por 'None':")
print(df.head()) # Imprime as primeiras linhas para verificar
```

```python
# Substituindo "?" por "NaN" nas colunas num√©ricas e convertendo para float**
# Identificando as colunas num√©ricas com valores faltantes
numeric_columns_with_missing_values = ['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price']

# Convertendo as colunas num√©ricas para float e substituindo "?" por None
for column in numeric_columns_with_missing_values:
  df[column] = pd.to_numeric(df[column], errors='coerce')

print(df) #Imprime dataframe

```

```python
# Limpando os valores faltantes da base de dados**
# Verificando a quantidade de valores faltantes em cada coluna
df.isnull().sum()
```

```python
 # Usando describe para substituir os valores faltantes pela m√©dia**
describe = df.describe()

print(describe)
```

```python
# **6. Substituindo os valores faltantes pela m√©dia de cada coluna**
for column in numeric_columns_with_missing_values:
  df[column] = df[column].fillna(describe.loc['mean', column])

print(df)
```

```python
# 1. Gr√°fico de Dispers√£o (Scatterplot) DarkMode
plt.figure(figsize=(15, 5.5))
plt.style.use('dark_background')  # Define o modo escuro
sns.scatterplot(data=df, x="horsepower", y="price", hue="fuel-type", palette="magma", marker="o", s=50, alpha=0.8)
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)
plt.gca().spines["left"].set_visible(False)
plt.ylabel("Price (Pre√ßo)")
plt.yticks(rotation=45)
plt.xlabel("Horsepower (Cavalos de Pot√™ncia)")
plt.xticks(range(50, 301, 25))
plt.grid(axis="y", alpha=0.5)
plt.title("Rela√ß√£o entre Pre√ßo e Pot√™ncia do Motor (com Tipo de Combust√≠vel)")
plt.legend(title='Fuel Type', loc='upper left')
plt.show()
```

```python
# 2. Boxplot Darkmode
plt.figure(figsize=(7, 6))
sns.boxplot(data=df, x="fuel-type", y="price", palette="coolwarm", showmeans=True, meanprops={"marker": "o", "markeredgecolor": "black", "markersize": 8})
plt.ylabel("Price (Pre√ßo)")
plt.yticks(rotation=45)
plt.xlabel("Fuel Type (Tipo de Combust√≠vel)")
plt.title("Compara√ß√£o de Pre√ßo por Tipo de Combust√≠vel")
plt.show()
```

```python
# 3. Histograma Darkmode
sns.displot(data=df, x="engine-size", height=6, aspect=2, kde=True, color="cyan", rug=True)
plt.title("Distribui√ß√£o do Tamanho do Motor")
plt.xlabel("Engine Size (Tamanho do Motor)")
plt.ylabel("Frequ√™ncia")
plt.show()
```
