```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = "https://arvores.fearp.usp.br/index"
```

```python
html = requests.get(url)
```

```python
html.text
```

```python
soup = BeautifulSoup(html.text, 'html.parser')
```

```python
print( soup.prettify() )
```

```python
# Encontre todas as divs que possuem a classe "card"
cards = soup.find_all('div', class_='card')
```

```python
# Exemplo de saída
print(cards[0].prettify())
```

```python
lista_nomes = []
lista_cientifico = []
lista_imagens = []
lista_codigo = []
lista_porte = []
lista_latitude = []
lista_longitude = []

# Extraindo informações de cada card
for card in cards:
  lista_nomes.append(card.find('b').get_text())
  lista_cientifico.append(card.find('small').find('i').get_text().replace('(', '').replace(')', ''))
  lista_imagens.append(card.find('img')['src'])
  lista_codigo.append( card.find('small').get_text().split("Código: ")[1].split("\n")[0] )
  lista_porte.append(card.find('small').get_text().split("Código: ")[1].split("\n")[1].split('Porte:')[1])
  coord = card.find('a', class_='btn btn-sm btn-primary')['href'].split('https://www.google.com.br/maps/search/')[1]
  lista_latitude.append(coord.split(',')[0])
  lista_longitude.append(coord.split(',')[1])

df = pd.DataFrame({'img': lista_imagens,
                   'nome': lista_nomes,
                   'nome_cientifico': lista_cientifico,
                   'codigo': lista_codigo,
                   'porte': lista_porte,
                   'latitude': lista_latitude,
                   'longitude': lista_longitude
                   })
```

## Varendo todas as páginas

```python
url = "https://arvores.fearp.usp.br/index?page="
```

```python
lista_nomes = []
lista_cientifico = []
lista_imagens = []
lista_codigo = []
lista_porte = []
lista_latitude = []
lista_longitude = []

# Varredura de todas as páginas
for idx in range(1,5+1):
  print("Lendo da página: " + url + str(idx))
  html = requests.get(url + str(idx))
  soup = BeautifulSoup(html.text, 'html.parser')
  # Encontre todas as divs que possuem a classe "card"
  cards = soup.find_all('div', class_='card')
  # Extraindo informações de cada card
  for card in cards:
    lista_nomes.append(card.find('b').get_text())
    lista_cientifico.append(card.find('small').find('i').get_text().replace('(', '').replace(')', ''))
    lista_imagens.append(card.find('img')['src'])
    lista_codigo.append( card.find('small').get_text().split("Código: ")[1].split("\n")[0] )
    lista_porte.append(card.find('small').get_text().split("Código: ")[1].split("\n")[1].split('Porte:')[1])
    coord = card.find('a', class_='btn btn-sm btn-primary')['href'].split('https://www.google.com.br/maps/search/')[1]
    lista_latitude.append(coord.split(',')[0])
    lista_longitude.append(coord.split(',')[1])

  df = pd.DataFrame({'img': lista_imagens,
                    'nome': lista_nomes,
                    'nome_cientifico': lista_cientifico,
                    'codigo': lista_codigo,
                    'porte': lista_porte,
                    'latitude': lista_latitude,
                    'longitude': lista_longitude
                    })
df.to_csv('arvores.csv', index=False)
df
```
