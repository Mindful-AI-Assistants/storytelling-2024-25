```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = "https://books.toscrape.com/"

html = requests.get(url)


# Parse HTML Code
soup = BeautifulSoup(html.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
# Encontre todas as divs que possuem a classe "card"
cards = soup.find_all('article', class_='product_pod')
```

### Escalar a solução para todas as páginas

```python
idx = 1

url = f'https://books.toscrape.com/catalogue/page-{idx}.html'
url
```

```python
df = pd.DataFrame({'titulo': [], 'thumb': [], 'preco': [], 'nota': [], 'estoque': []})
df
```

```python
for idx in range(1,51):
  url = f'https://books.toscrape.com/catalogue/page-{idx}.html'
  cards = soup.find_all('article', class_='product_pod')

  titulos = [card.find('h3').find('a')['title'] for card in cards]
  thumbs = ['https://books.toscrape.com/' + card.find('img')['src'] for card in cards]
  precos = [card.find('p', class_="price_color").text for card in cards ]
  notas = [card.find('p')['class'][1] for card in cards]
  estoque = [card.find('p', class_="instock availability").text.strip() for card in cards]

  aux = pd.DataFrame({'titulo': titulos, 'thumb': thumbs, 'preco': precos, 'nota': notas, 'estoque': estoque})
  df = pd.concat([df, aux])
```

```python
df
```

```python
df['preco'] = df['preco'].str.replace('£', '').astype(float)
df['nota'] = df['nota'].map({'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5})
df
```

```python
df.to_csv('livros.csv', index=False)
```
