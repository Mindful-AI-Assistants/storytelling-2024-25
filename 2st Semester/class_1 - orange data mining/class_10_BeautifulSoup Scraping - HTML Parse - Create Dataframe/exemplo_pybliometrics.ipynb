```python
import pybliometrics
from pybliometrics.scopus import ScopusSearch
```

```python
pybliometrics.scopus.init()  # read API keys
```

```python
s = ScopusSearch("TITLE-ABS-KEY ( \"machine learning\" ) AND ( PUBYEAR = 2025  )", verbose=True)
```

```python
import pandas as pd

df = pd.DataFrame(s.results)
df
```

```python
s = ScopusSearch("ORCID ( 0000-0002-8792-6054 )", verbose=True)

df = pd.DataFrame(s.results)
df
```
