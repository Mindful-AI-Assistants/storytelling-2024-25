```python
url ='https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data'
df = pd.read_csv(url, header=None)
df
```

# 2) O que seria uma regressão linear?

Imagine que já conhecemos a equação da reta:

$$y = mx+b $$

```python
import numpy as np
import pandas as pd

m = 2
b = 3

x = np.array([0, 1, 2, 3, 4, 5])
y = m*x + b


```

```python
lm_df = pd.DataFrame({'x':x, 'y':y})
lm_df
```

```python
%pip install scikit-learn

from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(x.reshape(-1, 1), y.reshape(-1, 1))
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
lm.predict([[3.5]])
```

```python
# Usando diretamento o DataFrame
xx = lm_df['x'].values.reshape(-1, 1)
yy = lm_df['y'].values.reshape(-1, 1)

lm = LinearRegression()
lm.fit(xx, yy)

lm.predict([[3.5]])
```

```python
x_interp = np.linspace(0,5,100)

y_hat = lm.predict( x_interp.reshape(-1,1) )


plt.plot(x_interp,y_hat,'r--')
plt.plot(xx,yy,'k*')
plt.show()
```

```python
lm.score(xx,yy)
```

### Regressão polinomial

* Quadrático
$$
\hat{y} = a + b_1 X + b_2 X^2
$$

* Cúbico
$$
\hat{y} = a + b_1 X + b_2 X^2 + b_3 X^3
$$

* Alta ordem
$$
\hat{y} = a + b_1 X + b_2 X^2 + b_3 X^3 + \cdots
$$

```python
x = df['highway-mpg']
y = df['price']
```

```python
f = np.polyfit(x,y,1)
f
```

```python
p = np.poly1d(f)
print(p)
```

```python
f = np.polyfit(x,y,4)
f
```

```python
p = np.poly1d(f)
print(p)
```

```python
xx = np.linspace( x.min(), x.max(), 100  )
yy = p(xx)

plt.plot(x,y,'k.',
         xx,yy,'r');
```

```python
y_hat = p(x)
```

```python
p(47.5)
```

```python
r2_score(y, y_hat)
```
