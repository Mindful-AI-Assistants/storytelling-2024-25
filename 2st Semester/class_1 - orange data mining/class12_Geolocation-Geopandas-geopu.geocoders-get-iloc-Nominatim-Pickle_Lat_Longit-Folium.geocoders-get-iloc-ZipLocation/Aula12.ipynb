```python
def prettycep(cep):
  cep = str(cep)
  if len(cep) == 7:
    cep = '0' + cep
  if len(cep) == 8:
    return cep[:5] + '-' + cep[5:]
  return cep
```

```python
CEP = prettycep(df.iloc[0]['CEP'])
out = brazilcep.get_address_from_cep(CEP)

print( out['street'] + ' , ' + out['district'], ' , ' + out['cep'] + ' , ' + out['uf'] )

```

```python
from numpy import nan
from tqdm.notebook import tqdm


from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="geolocation")

latitudes = []
longitudes = []

for index in tqdm(range(len(locais_unicos))):
  try:
    #CEP = prettycep(locais_unicos.iloc[index]['CEP'])
    #out = brazilcep.get_address_from_cep(CEP)
    #endereco = out['street'] + ' , ' + out['district'] + ' , ' + out['cep'] + ' , ' + out['uf']
    endereco = locais_unicos.iloc[index]['LOCAL_VOTACAO'] + ' , ' + locais_unicos.iloc[index]['ENDERECO'] + ' , ' + locais_unicos.iloc[index]['BAIRRO_LOCAL_VOT'] + ' , SÃO PAULO - SP'
    location = geolocator.geocode(endereco)
    if location:
      latitudes.append(location.latitude)
      longitudes.append(location.longitude)
  except:
    pass
```

```python
import pickle

with open('latitudes.pickle', 'wb') as f:
  pickle.dump(latitudes, f)

with open('longitudes.pickle', 'wb') as f:
  pickle.dump(longitudes, f)
```

```python
import folium
```

```python
map = folium.Map(location=[-23.550520, -46.633308], zoom_start=10)

for lat, lng in zip(latitudes, longitudes):
  folium.Marker([lat, lng]).add_to(map)

map
```
