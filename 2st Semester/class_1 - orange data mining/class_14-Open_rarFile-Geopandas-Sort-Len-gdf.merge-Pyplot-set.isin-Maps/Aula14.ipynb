```python
# Contagem de quantas sees temos!
CONTAGEM_SECOES = df_SP.groupby('ZONA')['NUM_SECAO'].count().reset_index()
CONTAGEM_SECOES.rename(columns={'NUM_SECAO': 'CONTAGEM'}, inplace=True)
CONTAGEM_SECOES
```

```python
gdf_merged = gdf.merge(right=CONTAGEM_SECOES, left_on='Name', right_on='ZONA')
gdf_merged
```

```python
gdf_merged.plot('CONTAGEM', legend=True)
```

```python
import matplotlib.pyplot as plt

# Plotando o mapa com a contagem
fig, ax = plt.subplots(figsize=(10, 8))
gdf_merged.plot('CONTAGEM', legend=True, ax=ax, cmap='Reds')
gdf_merged.apply(lambda x: ax.annotate(text=x['ZONA'], xy=x.geometry.centroid.coords[0], ha='center', fontsize=6), axis=1);

# Melhorando o grfico
ax.set_title('Distribuio de Contagem das Sees por Zona', fontsize=16)
ax.set_xlabel('Longitude', fontsize=12)
ax.set_ylabel('Latitude', fontsize=12)

# Exibir o grfico
plt.show()
```

```python
df2 = pd.read_csv('eleicoes.csv')
df2
```

```python
filtro = df2['NM_MUNICIPIO'] == 'SO PAULO'
filtro2 = df2['DS_CARGO_PERGUNTA'] == 'Presidente'

df2[filtro & filtro2]['NM_VOTAVEL'].unique()
```

```python
df_PRESIDENTE = df2[filtro & filtro2].groupby(['NR_ZONA', 'NM_VOTAVEL'])['QT_VOTOS'].sum().reset_index()
df_PRESIDENTE
```

```python
dfs = {}

for candidato in df2[filtro & filtro2]['NM_VOTAVEL'].unique():
  aux = df_PRESIDENTE['NM_VOTAVEL'] == candidato
  dfs[candidato] = df_PRESIDENTE[aux]
```

```python
dfs
```

```python
gdf_aux = gdf.merge(right=dfs['PADRE KELMON'], left_on='Name', right_on='NR_ZONA')
gdf_aux
```

```python
import matplotlib.pyplot as plt

# Plotando o mapa com a contagem
fig, ax = plt.subplots(figsize=(10, 8))
gdf_aux.plot('QT_VOTOS', legend=True, ax=ax, cmap='Blues')
gdf_aux.apply(lambda x: ax.annotate(text=x['NR_ZONA'], xy=x.geometry.centroid.coords[0], ha='center', fontsize=6), axis=1);

# Melhorando o grfico
ax.set_title('Votos no Padre Kelmon por Zona', fontsize=16)
ax.set_xlabel('Longitude', fontsize=12)
ax.set_ylabel('Latitude', fontsize=12)

# Exibir o grfico
plt.show()
```

As cores esto enviesadas pela quantidade se sees em cada zona eleitoral. Precisamos fazer um percentual para a anlise fazer mais sentido.

```python
TOTAL_GERAL = df_PRESIDENTE = df2[filtro & filtro2].groupby(['NR_ZONA'])['QT_VOTOS'].sum().reset_index()
TOTAL_GERAL.rename(columns={'QT_VOTOS': 'TOTAL_GERAL'}, inplace=True)
TOTAL_GERAL
```

```python
candidato = 'PADRE KELMON'

gdf_aux = gdf.merge(right=TOTAL_GERAL, left_on='Name', right_on='NR_ZONA')
gdf_aux2 = gdf_aux.merge(right=dfs[candidato], left_on='Name', right_on='NR_ZONA')
gdf_aux2['PERCENTUAL'] = gdf_aux2['QT_VOTOS'] / gdf_aux2['TOTAL_GERAL'] * 100


# Plotando o mapa com a contagem
fig, ax = plt.subplots(figsize=(10, 8))
gdf_aux2.plot('PERCENTUAL', legend=True, ax=ax, cmap='Reds')
gdf_aux2.apply(lambda x: ax.annotate(text=x['Name'], xy=x.geometry.centroid.coords[0], ha='center', fontsize=6), axis=1);

# Melhorando o grfico
ax.set_title(f'Votos no {candidato} por Zona', fontsize=16)
ax.set_xlabel('Longitude', fontsize=12)
ax.set_ylabel('Latitude', fontsize=12)

# Exibir o grfico
plt.show()
```
